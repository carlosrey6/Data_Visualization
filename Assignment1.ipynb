```python
# 1. Loading Dataset and examinning the dimentions. 

import pandas as pd     #Importing pandas library

dataset = pd.read_excel('flight_price.xlsx')    #Reading a file xlsx in pandas. 


rows, columns = dataset.shape   #Counting the rows and columns in the data set using shape property

print("Total rows: ", rows, "Total columns: ", columns) #Pinting the total of rows and columns 
```

PART 2 - GOOGLE PLAYSTORE

```python
#9 Loading Google Playstore Dataset and Examinning dimentions (rows, columns) 


dataset = pd.read_csv('googleplaystore.csv') #Loading the dataset, i dont have to import the library again because i already have the library imported by the previous exercise.

rows, columns = dataset.shape   #Counting the rows and columns in the data set using shape property
print("The Google Playstore Dataset has a total of ", rows, " rows and a total of ", columns, " columns") #Pinting the total of rows and columns
```

```python
#10 Variation of ratings by categories. Using a boxplot

plt.figure(figsize=(12, 8)) #Setting figure size 
sns.boxplot(x='Category', y='Rating', data=dataset) #Creating the boxplot to visualize the data, with 'Category' = x-axis, and  'Rating' on the y-axis.
plt.title('App Ratings by Category') #Setting the tittle of the graph
plt.xticks(rotation=90) #Im using this to rotate the x-axis labels by 90 degrees to improve the readability. 
plt.show()  #Displaying the graph 
```

```python
#11 Checking missing values: 

missing_values = dataset.isnull().sum() #Creating a variable using isnull() method to identify missing values in each column, and using sum() method to count the number of missing values.
print('Misssing values: ', missing_values)   #Printing the variable 
```

```python
#12 Relationship between the size of an app and its rating, using a scatterplot. 

plt.figure(figsize=(14, 10))  #Setting the size of the new figure
sns.set(style="whitegrid")  #Setting the style of the theme 
sns.scatterplot(x='Size', y='Rating', data=dataset)  #Creating the scatterplot and setting the x and y axis names. 
plt.title('Relationship between App Size and Rating')   #Setting the title of the graph
plt.xticks([])  #Removing the tick marks and labels on the x-axis, just because there are a lot of data, 
plt.show()
```

```python
#13. How does the type of app affect its price, using a bar chart. 

# Converting the price column to numeric values, eliminating $ sign: 
dataset['Price_Dollars'] = pd.to_numeric(dataset['Price'].str.replace('$', ''), errors='coerce')    # The errors='coerce' option assigns NaN to values that cannot be converted.

# Creating the figure an setting the size. 
plt.figure(figsize=(10, 6))
# Creatting the bar plot. On the x-axis, we want the categories, on the y-axis, we want the prices in dollars. 
sns.barplot(x='Category', y='Price_Dollars', data=dataset, ci=None, estimator='mean', order=dataset.groupby('Category')['Price_Dollars'].mean().sort_values(ascending=False).index)

# Setting the labels and the title:
plt.title('Average Prices by App Type')
plt.xlabel('App Category')
plt.ylabel('Average Price')

# Customizing the xticks
plt.xticks(rotation=90)  # Rotar las etiquetas del eje x para mayor claridad

# Displaying the plot
plt.show()  
```

```python
#14. Top 10 most popular apps in the dataset. Using a frequency table. 

# Groupping by 'App' and sum the installs in a new variable 'app_installs' 
app_installs = dataset.groupby('App')['Installs'].sum().reset_index()

# Sortting by installs in descending order
sorted_apps = app_installs.sort_values(by='Installs', ascending=False)

# Displaying the top 10 apps
top_10_apps = sorted_apps.head(10)
print(top_10_apps)
```

```python
#15. A company wants to launch a new app...analyze the Google Playstore dataset to identify the most popular app categories. 


# Calculating total installations per category
installs_category = dataset.groupby('Category')['Installs'].sum().sort_values(ascending=False)

# Visualizing the top 10 most popular categories in a bar chart
plt.figure(figsize=(10, 6))
installs_category.head(10).plot(kind='bar', color='orange')
plt.title('Top 10 App Categories by Total Installs')
plt.xlabel('Category')
plt.ylabel('Total Installs')
plt.xticks(rotation= 90)
plt.ticklabel_format(style='plain', axis='y')
plt.show()

# Printing the most popular categories
top_categories = installs_category.head(10).index
print("Top 10 App Categories by Total Installs")
for category in top_categories:
    print(f"- {category}")
```

```python
#16.A mobile app development company wants to analyze the Google Playstore dataset to identify the most successful app developers

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the dataset
df = pd.read_csv('googleplaystore.csv')

# Cleanning 'Installs' column to remove non-numeric values
df['Installs'] = df['Installs'].str.replace(',', '').str.replace('+', '')

# Converting 'Installs' to numeric format
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')

# Filtering only free apps (Type == 'Free')
df_free_apps = df[df['Type'] == 'Free']

# Group by developer and calculate metrics
numeric_cols = ['Installs', 'Rating', 'Reviews']
df_free_apps[numeric_cols] = df_free_apps[numeric_cols].apply(pd.to_numeric, errors='coerce')
developer_stats = df_free_apps.groupby('App')[numeric_cols].mean()
developer_stats['Number of Apps'] = df_free_apps.groupby('App').size()

# Filtering developers with more than one app
developer_stats = developer_stats[developer_stats['Number of Apps'] > 1]

#Sorting developers by average installs
top_developers = developer_stats.sort_values(by='Installs', ascending=False).head(10)

# Displaying metrics for the top 10 developers
print(top_developers)

# Data visualizations
plt.figure(figsize=(12, 8))

# Bar chart for average installs per developer
sns.barplot(x=top_developers.index, y=top_developers['Installs'])
plt.title('Top 10 Developers by Average Installs')
plt.xlabel('Developers')
plt.ylabel('Average Installs')
plt.xticks(rotation=45, ha='right')
plt.show()

# Bar chart for average ratings per developer
plt.figure(figsize=(12, 8))
sns.barplot(x=top_developers.index, y=top_developers['Rating'])
plt.title('Top 10 Developers by Average Rating')
plt.xlabel('Developers')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.show()

# Bar chart for average reviews per developer
plt.figure(figsize=(12, 8))
sns.barplot(x=top_developers.index, y=top_developers['Reviews'])
plt.title('Top 10 Developers by Average Reviews')
plt.xlabel('Developers')
plt.ylabel('Average Reviews')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading the dataset
df = pd.read_csv('googleplaystore.csv')

# Try to infer the date format automatically
df['Last Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce', infer_datetime_format=True)

# Scatter plot de Installs vs. Last Updated
plt.figure(figsize=(12, 6))
plt.scatter(df['Last Updated'], df['Installs'], alpha=0.5)
plt.title('Installs vs. Last Updated')
plt.xlabel('Last Updated')
plt.ylabel('Installs')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# Histogram of Ratings
plt.figure(figsize=(8, 6))
plt.hist(df['Rating'].dropna(), bins=20, edgecolor='black')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

# Scatter plot of Reviews vs. Ratings
plt.figure(figsize=(8, 6))
plt.scatter(df['Reviews'], df['Rating'], alpha=0.5)
plt.title('Reviews vs. Ratings')
plt.xlabel('Reviews')
plt.ylabel('Rating')
plt.xscale('log')  # Usar escala logartmica para manejar valores extremos
plt.show()

# Bar chart of Category-wise Analysis
category_counts = df['Category'].value_counts()
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar')
plt.title('Category-wise Analysis')
plt.xlabel('Category')
plt.ylabel('Number of Apps')
plt.xticks(rotation=90)
plt.show()
```
